```python
from keras.models import Model
from keras.layers import Input,Dense,Multiply,Lambda
from keras import backend as K
from keras import losses
from keras.optimizers import Adam
from keras.utils import to_categorical
from keras.models import load_model
import gym
from collections import deque
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
env=gym.make('CartPole-v0')
action_space_dim=env.action_space.n
state_space_dim=4
```

## Training Algorithm

Memory-Replay

* Initialize the memory $D$
* Initialize the action-value network $Q$ with random weights
* For episode = 1, $M$ do
  * For $t$, $T$ do
    * With probability $\epsilon$ select a random action $a_t$, otherwise select $a_t = \mathrm{argmax}_a Q(s,a)$
    * Execute action $a_t$ in simulator and observe reward $r_{t+1}$ and new state $s_{t+1}$
    * Store transition <s_t, a_t, r_{t+1}, s_{t+1}> in memory $D$
    * Sample random mini-batch from $D$: <s_j, a_j, r_j, s'_j>
    * Set $\hat{Q}_j = r_j$ if the episode ends at $j+1$, otherwise set $\hat{Q}_j = r_j + \gamma \max_{a'}{Q(s'_j, a')}$
    * Make a gradient descent step with loss $(\hat{Q}_j - Q(s_j, a_j))^2$


```python
def Qnet(learning_rate=0.001):
    x=Input(shape=(state_space_dim,),name='input_1')
    f1=Dense(24,activation='relu',name='dense_1')(x)
    f2=Dense(24,activation='relu',name='dense_2')(f1)
    y=Dense(action_space_dim,activation='linear',name='q_estimate')(f2)
    model=Model(inputs=x,outputs=y)
    model.compile(optimizer=Adam(lr=learning_rate),loss='mse')
    return model
```

```python
qnet=Qnet(0.001)
qnet.summary()
```

```python
memory=deque(maxlen=2000)
for episode in range(20):
    state=env.reset()
    for t in range(250):
        #env.render()
        action=env.action_space.sample()
        next_state,reward,done,_=env.step(action)
        memory.append((state,action,reward,next_state,done))
        state=next_state
        if done:
            break
#env.close()
```

```python
gamma=0.99
batch_size=32
max_training_episodes=1000
decay_rate=0.85
```

```python
def get_batch(memory,batch_size=16):
    ids=np.random.choice(np.arange(len(memory)),size=batch_size,replace=False)
    batch=[memory[i] for i in ids]
    states=np.array([e[0] for e in batch])
    actions=np.array([e[1] for e in batch])
    rewards=np.array([e[2] for e in batch])
    next_states=np.array([e[3] for e in batch])
    statuses=np.array([e[4] for e in batch])
    return states,actions,rewards,next_states,statuses
```

```python
finished_after=[]
average_finish_times_over_past_100_episodes=[]
ntimes_finish_time_over_200=0
accumulated_finish_time=0
epsilons=[]
loss_history=[]
for episode in range(max_training_episodes):
    epsilon=max(0.01,min(1,1-decay_rate*np.log10((episode+1)/25)))
    epsilons.append(epsilon)
    state=env.reset()
    for t in range(250):
        #env.render()
        if np.random.random()<epsilon:
            action=env.action_space.sample()
        else:
            q_pred=qnet.predict(np.array([state]))
            action=np.argmax(q_pred)
        next_state,reward,done,_=env.step(action)
        if done:
            if t<199:
                reward=-5
            else:
                reward=0
        memory.append((state,action,reward,next_state,done))
        state=next_state
        
        states,actions,rewards,next_states,statuses=get_batch(memory,batch_size)
        q_pred_next_states=qnet.predict(next_states)
        targets=rewards+gamma*np.amax(q_pred_next_states,axis=1)
        done_statuses=(statuses==1)
        targets[done_statuses]=rewards[done_statuses]
        targets_=qnet.predict(states)
        for i in range(batch_size):
            targets_[i][actions[i]]=targets[i]
        l=qnet.train_on_batch(states,targets_)
        loss_history.append(l)
        if done:
            print("Epsiode",episode+1,"ended at time step",t+1)
            finished_after.append(t+1)
            accumulated_finish_time=accumulated_finish_time+(t+1)
            if len(average_finish_times_over_past_100_episodes)<99:
                average_finish_times_over_past_100_episodes.append(t+1)
            else:
                average_finish_times_over_past_100_episodes.append(accumulated_finish_time/100)
                accumulated_finish_time=accumulated_finish_time-average_finish_times_over_past_100_episodes[-100]
            if (t+1)>=200:
                ntimes_finish_time_over_200=ntimes_finish_time_over_200+1
            else:
                ntimes_finish_time_over_200=0
            break
    if ntimes_finish_time_over_200>=50:
        break
#env.close()
file_name='qnet.h5'
if os.path.isfile(file_name):
    os.remove(file_name)
qnet.save('qnet.h5')
```
